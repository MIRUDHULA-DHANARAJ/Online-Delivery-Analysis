```python
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())
df['Customer_Age']
```

```python
df['Customer_Gender'].value_counts()
```

```python
df['Customer_Gender'] = df['Customer_Gender'].fillna(df['Customer_Gender'].mode()[0])
```

```python
df['Customer_Gender'].isnull().sum()
```

```python
cat_col = ['City','Area','Cuisine_Type','Payment_Mode']  
```

```python
for col in cat_col:
    if not df[col].mode().empty:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df[cat_col].isnull().sum()
```

```python
for col in cat_col:
    df[col]=df[col].str.strip().str.title()
```

```python
df.loc[df['Order_Status'] == 'Delivered', 'Cancellation_Reason'] = 'Not Cancelled'
```

```python
df.loc[df['Order_Status'] == 'Cancelled', 'Delivery_Rating'] = np.nan
```

```python
df.loc[(df['Order_Value'].isnull()) & (df['Order_Status']=='Cancelled'), 'Order_Value'] = 0
```

```python
median_order_value = df[df['Order_Status']=='Delivered']['Order_Value'].median()
df.loc[(df['Order_Value'].isnull()) & (df['Order_Status']=='Delivered'), 'Order_Value'] = median_order_value
```

```python
df['Discount_Applied'] = df['Discount_Applied'].fillna(0)
```

```python
df.loc[df['Order_Status']=='Delivered', 'Final_Amount'] = df['Order_Value'] - df['Discount_Applied']
```

```python
df.loc[df['Order_Status']=='Cancelled', 'Final_Amount'] = 0
```

```python
df['Distance_km']=df['Distance_km'].fillna(df['Distance_km'].median())
```

```python
df['Delivery_Time_Min'] = df.groupby('City')['Delivery_Time_Min'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df['Delivery_Time_Min'].describe()
```

```python
for col in ['Delivery_Time_Min','Order_Value']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df[col] = np.where(df[col]<lower, lower, np.where(df[col]>upper, upper, df[col]))
```

```python
df['Order_Value'].describe()
```

```python
rating_col = ['Delivery_Rating', 'Restaurant_Rating']
for col in rating_col:
    median_val = df[col].median()
    df.loc[(df[col]<1) | (df[col]>5), col] = median_val
    df[col] = df[col].fillna(median_val)
```

```python
df.loc[df['Profit_Margin'] < 0, 'Profit_Margin'] = 0
df.loc[df['Profit_Margin'] > 1, 'Profit_Margin'] = 1
df['Profit_Margin'] = df['Profit_Margin'].fillna(df['Profit_Margin'].median())
```

```python
df['Profit_Margin_Percent'] = df['Profit_Margin'] * 100
```

```python
df['Peak_Hour'] = df['Peak_Hour'].map({True:1, False:0})
df['Peak_Hour'] = df['Peak_Hour'].fillna(df['Peak_Hour'].mode()[0])
```

```python
df['Order_Date'] = df['Order_Date'].astype(str).str.strip()
```

```python
a = df['Order_Date'].str.contains("/")
b = df['Order_Date'].str.contains("-")
```

```python
df.loc[a, 'Order_Date'] = pd.to_datetime(df.loc[a,'Order_Date'], format='%m/%d/%Y', errors='coerce')
df.loc[b, 'Order_Date'] = pd.to_datetime(df.loc[b,'Order_Date'], format='%d-%m-%Y', errors='coerce')
```

```python
df['Order_Date'] = df['Order_Date'].fillna(df['Order_Date'].mode()[0])
```

```python
df.to_csv("../data/processed/cleaned_data.csv", index=False)
```
