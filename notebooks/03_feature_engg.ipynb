```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Day_Name'] = df['Order_Date'].dt.day_name()

```

```python
df['Day_Type'] = 'Weekday'
df.loc[df['Day_Name'].isin(['Saturday','Sunday']), 'Day_Type'] = 'Weekend'
```

```python
df['Profit_Margin_Percent'] = df['Profit_Margin'] * 100
```

```python
df['Delivery_Performance'] = 'Moderate'

df.loc[df['Delivery_Time_Min'] < 30, 'Delivery_Performance'] = 'Fast'
df.loc[df['Delivery_Time_Min'] > 60, 'Delivery_Performance'] = 'Delayed'
```

```python
df['Age_Group'] = 'Adult'

df.loc[df['Customer_Age'] < 25, 'Age_Group'] = 'Young'
df.loc[df['Customer_Age'] > 40, 'Age_Group'] = 'Senior'
```

```python
df.to_csv("../data/processed/feature_engineered_data.csv", index=False)
```
