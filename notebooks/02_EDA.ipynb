```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Order_Value'], bins=30, kde=True)
plt.title('Distribution of Order Value')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['Delivery_Time_Min'], bins=40)
plt.title("Distribution of Delivery Time")
plt.xlabel("Delivery Time")
plt.ylabel("Frequency")
plt.show()
```

```python
df['City'].value_counts().plot(kind='bar')
plt.title("City-wise Orders")
plt.show()
```

```python
df['Cuisine_Type'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Cuisines")
plt.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Day_Name'] = df['Order_Date'].dt.day_name()

```

```python
df['Day_Type'] = 'Weekday'
df.loc[df['Day_Name'].isin(['Saturday','Sunday']), 'Day_Type'] = 'Weekend'
```

```python
df['Day_Type'].value_counts().plot(kind='bar')
plt.title('Weekend vs Weekday Orders')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.scatter(df['Distance_km'], df['Delivery_Time_Min'])
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Time")
plt.title("Distance vs Delivery Time")
plt.show()
```

```python
cancel_df = df[df['Order_Status']=='Cancelled']
cancel_df['Cancellation_Reason'].value_counts().plot(kind='bar')
plt.title("Cancellation Reasons")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(11,8))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
